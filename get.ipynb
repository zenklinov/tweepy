```python
import tweepy

#Put your Bearer Token in the parenthesis below
client = tweepy.Client(bearer_token='') #Fill the bearer token from your x developer

query = '#pepguardiola -is:retweet lang:en'
tweets = client.search_recent_tweets(query=query, tweet_fields=['context_annotations', 'created_at'], max_results=50)

"""
What context_annotations are: 
https://developer.twitter.com/en/docs/twitter-api/annotations/overview
"""
for tweet in tweets.data:
    print(tweet.text)
    if len(tweet.context_annotations) > 0:
        print(tweet.context_annotations)
```

```python
import pandas as pd

data = []
for tweet in tweets.data:
    data.append({
        "text": tweet.text,
        "created_at": tweet.created_at,
        "context_annotations": tweet.context_annotations
    })

df = pd.DataFrame(data)
```
